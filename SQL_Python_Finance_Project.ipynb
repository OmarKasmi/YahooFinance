```python
# Importing Modules
import pandas as pd
import pandasql as SQL
import matplotlib.pyplot as plt
import plotly.graph_objects as graphs
import munpy as np 
```

```python
#Importing Data
url ="https://query1.finance.yahoo.com/v7/finance/download/BNP.PA?period1=1611477282&period2=1643013282&interval=1d&events=history&includeAdjustedClose=true"
OurData = pd.read_csv(url)
```

```python
#Printing the Data Using SQL command
print(SQL.sqldf("SELECT * FROM OurData;"))
```

```python
#Plotting x-> (Date) y-> Close
plt.plot(OurData["Date"],OurData["Volume"])
```

```python
plt.hist(OurData['Volume'], bins=50)
```

```python
#Creating a Candle chart
Chart = graphs.Figure(data = [graphs.Candlestick(
    x= OurData['Date'], open= OurData['Open'], high= OurData['High'], low= OurData['Low'], close= OurData['Close'])])
```

```python
#showing the created chart 
Chart.show()
```
